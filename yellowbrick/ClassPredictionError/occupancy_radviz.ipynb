```python
# Load the classification data set
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('data/data/occupancy/occupancy.csv')
from yellowbrick.style.palettes import color_palette

# Specify the features of interest and the classes of the target
features = ["temperature", "relative humidity", "light", "C02", "humidity"]
classes = ['unoccupied', 'occupied']

# Extract the numpy arrays from the data frame
X = data[features].as_matrix()
y = data.occupancy.as_matrix()
# Import the visualizer
from yellowbrick.features import RadViz
```

```python
# Instantiate the visualizer
visualizer = RadViz(classes=classes, features=features, colormap='flag')

visualizer.fit(X, y)      # Fit the data to the visualizer
visualizer.transform(X)   # Transform the data
visualizer.poof()
plt.get_cmap('flag')

plt.show()
```

```python
from yellowbrick.features import FeatureImportances
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston, load_breast_cancer, load_wine
from sklearn.ensemble import RandomForestRegressor
import numpy as np

df = pd.read_csv('data/reddit_data.csv')
features = ['wordcount', 'charcount', 'vowels', 'consonants', 'syllable', 'polarity', 'subjectivity', 'noun_phrases', 'kincaid', 'flesch']

X = df[features]
y = df.gtavg

clf = RandomForestClassifier(n_estimators=190, n_jobs=-1, max_depth=5, max_features='log2', min_samples_leaf=1, min_samples_split=77)
viz = FeatureImportances(clf)
viz.fit(X, y)      # Fit the data to the visualizer
print(viz.feature_importances_)
viz.poof()
```
